```python
cnbc = load_dataset("csv", data_files = "cnbc_headlines.csv")["train"]
reuters = load_dataset("csv", data_files = "reuters_headlines.csv")["train"]
#notes = notes.train_test_split(test_size=0.2)
```

```python
notes = concatenate_datasets([cnbc, reuters])
```

```python
notes = notes.filter(lambda x: x["Headlines"] != None)
```

```python
notes = notes.train_test_split(test_size = 0.2)
```

```python
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-base") # from_pretrained
```

```python
#prefix = "summarize: "

def preprocess_function(examples):
    inputs = [doc for doc in examples["Description"]]
    model_inputs = tokenizer(inputs, max_length=1024, truncation=True)

    labels = tokenizer(text_target=examples["Headlines"], max_length=128, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# short description을 집어넣어서 headline을 가져오는 것
# language model 두개를 쌓을 것
# 요약하는 것 하나 + CLM으로 하나 둘 다 BART임
```

```python
tokenized_data = notes.map(preprocess_function, batched=True, num_proc=4, remove_columns=notes["train"].column_names)
```

```python
from transformers import DataCollatorForLanguageModeling, DataCollatorForSeq2Seq

tokenizer.pad_token = tokenizer.eos_token
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model= "facebook/bart-base")
```

```python
from transformers import AutoModelForCausalLM, TrainingArguments, Trainer, AutoModelForSeq2SeqLM

model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-base")
```

```python
import evaluate

rouge = evaluate.load("rouge")
```

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    result = rouge.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)

    prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in predictions]
    result["gen_len"] = np.mean(prediction_lens)

    return {k: round(v, 4) for k, v in result.items()}
```

```python
batch_size = 8

training_args = Seq2SeqTrainingArguments(
    output_dir="my_gen_head_model",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=4,
    predict_with_generate=True,
    fp16=True,
    push_to_hub=True,
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_data["train"],
    eval_dataset=tokenized_data["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
trainer.push_to_hub()
```

```python
from transformers import pipeline
model = AutoModelForSeq2SeqLM.from_pretrained("Pstman/my_gen_head_model")
tokenizer = AutoTokenizer.from_pretrained("Pstman/my_gen_head_model")

text = "TikTok has been in discussions with the UK government over the past few months to locate its headquarters in London, a source familiar with the matter said, as part of a strategy to distance itself from its Chinese ownership."
tokenizer(text)

# Tokenize our text
# If you want to run the code in Tensorflow, please remember to return the particular tensors as simply as using return_tensors = 'tf'
input_ids = tokenizer(text, return_tensors="pt").input_ids

# Generate the output (Here, we use beam search but you can also use any other strategy you like)
output = model.generate(
    input_ids,
    max_length=512,
    num_beams=5,
    early_stopping=True
)

text = tokenizer.decode(output[0], skip_special_tokens = True)
# Finally, we can print the generated summary
print(text)
```
